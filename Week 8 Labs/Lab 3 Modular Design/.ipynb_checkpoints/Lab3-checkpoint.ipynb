```
from pathlib import Path

code = '''
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mean(values):
    return sum(values)/len(values)
'''

Path('mymath.py').write_text(code)
```

```
import mymath

print("Add:", mymath.add(3, 4))
print("Subtract:", mymath.sub(9, 5))
print("Mean:", mymath.mean([1, 2, 3, 4]))
```
